```python
from Bio.SeqUtils import ProtParam
import pandas as pd
#contador de numero de peptideos em uma lista e composição de aminoácidos

peptide_data = []

with open('../data/raw/positive.txt') as reader: 
    for peptide in reader:
        aa_composition = ProtParam.ProteinAnalysis(peptide).get_amino_acids_percent()
        aa_composition['label'] = 1
        peptide_data.append(aa_composition)
        
with open('../data/raw/negative.txt') as reader: 
    for peptide in reader:
        aa_composition = ProtParam.ProteinAnalysis(peptide).get_amino_acids_percent()
        aa_composition['label'] = 0
        peptide_data.append(aa_composition)

#dataframe with pandas
        
df_peptides = pd.DataFrame(peptide_data)
df_peptides
```

```python
X = df_peptides.drop(['label'], axis=1)
y = df_peptides['label']
```

 ## Composição de aminoácidos diferencial

### Redução de dimensões 

```python
from sklearn.decomposition import PCA
from plotly import express as ex

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

df_peptides_pca = pd.DataFrame(X_pca, columns=['pca_1', 'pca_2'])
df_peptides_pca['label'] = y.astype(str)

ex.scatter(df_peptides_pca, x='pca_1', y='pca_2', color='label')
```

```python
from sklearn.metrics import silhouette_score

silhouette_score(X_pca, y)
```

```python
from umap import UMAP

umap = UMAP(n_components=2)
X_umap = umap.fit_transform(X)

df_peptides_umap = pd.DataFrame(X_umap, columns=['umap_1', 'umap_2'])
df_peptides_umap['label'] = y.astype(str)

ex.scatter(df_peptides_umap, x='umap_1', y='umap_2', color='label')
```

```python
silhouette_score(X_umap, y)
```

#### Árvore de decisão para treinamento do algoritmo
##### Teste dos algoritmos para que não aconteça um sobreajuste do mesmo e ele decore os resultados
##### Randomforest utilizado como padrão de modelo supervisionado
##### Para avaliar a confiabilidade do modelo, utilizar dados novos para comparar os resultados


```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn. linear_model import LogisticRegression
from sklearn.metrics import classification_report


X_train, X_test, y_train, y_test = train_test_split(X, y)

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

report = classification_report(y_test, y_pred)

print(report)
```

##### Recall -> avaliar os falsos negativos presentes na análise (quanto maior valor melhor)
##### precision -> avaliar os falsos positivos
##### f1score -> média harmonica entre recall e precision

```python
def predict_anticancer_peptice(peptide_sequence):
    aa_composition = ProtParam.ProteinAnalysis(peptide_sequence).get_amino_acids_percent()
    X = pd.DataFrame([aa_composition])
    return model.predict_proba(X)[0][1]

predict_anticancer_peptice('GIGYILDRQSTRKSWTRHFVKFGEGQDEAWRQWKGIYHLSMTTCYASFIA')
```
